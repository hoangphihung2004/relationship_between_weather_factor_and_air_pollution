### The Impact Of Weather On NO2:
- The correlation between temperature and NO2 is very low, almost negligible, indicating that temperature and NO2 do not have a linear relationship. But we can see from the trend chart that the temperature and NO2 trends are very similar. The higher the temperature, the higher the NO2.
- The correlation between rainfall and NO2 is very low, almost zero, indicating that rainfall and NO2 do not have a linear relationship. But we can see that in the early months of the year when rainfall is low, NO2 is high, and in the later months of the year when rainfall increases, NO2 decreases.
- The correlation between humidity and NO2 is very low, almost zero, indicating that humidity and NO2 do not have a linear relationship. But we can see that when humidity drops sharply, CO increases, especially in June, July, and August. But when humidity increases, CO decreases.
- The correlation between wind speed and NO2 is very low, almost zero, indicating that wind speed and NO2 do not have a linear relationship. But we can see that from January to September when the wind speed is low, NO2 tends to increase, but from October when the wind speed is high, NO2 decreases sharply.
=> Although there is no strong linear relationship between these weather factors and NO₂, temperature, rainfall, humidity, and wind speed may still indirectly influence NO₂ levels, especially when considering seasonal changes throughout the year.
