```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd  # Thêm pandas để tạo DataFrame

driver = webdriver.Chrome() 
driver.get("https://air.plumelabs.com/air-quality-in-Da%20Nang-6E4g")  # Replace with URL containing element

# Click the "Year" button to go to yearly data
year_button = driver.find_element(By.XPATH, "//li[@data-time-frame='year']")
year_button.click()
time.sleep(2)  # Wait for page to load data after clicking

# Scroll page down 100 pixels
driver.execute_script("window.scrollBy(0, 200);")
time.sleep(1)

# Find circle element by id
circle_element = driver.find_element(By.ID, "chart__point")
actions = ActionChains(driver)

# Click and hold the circle element
actions.click_and_hold(circle_element).perform()

# List to store date data and O3 index
data = []

# Set to store the dates that have appeared
unique_dates = set()

# Loop mouse movement and get date information and O3 index
for _ in range(485):  # Change the number of iterations as desired
    actions.move_by_offset(-1, 0).perform()

    # Get the current date from the date element
    date_element = driver.find_element(By.CSS_SELECTOR, "p#chart-date.report__lead")
    current_date = date_element.text

    # Only process if current date has not appeared before
    if current_date not in unique_dates:
        unique_dates.add(current_date)

        # Get O3 index
        o3_element = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[5]/div[2]/ul/li[4]/div[1]")
        o3_value = o3_element.text
        time.sleep(2)

        # Add data to list
        data.append([current_date, o3_value])

        # In ra màn hình ngày và chỉ số O3
        print(f"Ngày: {current_date}, O3: {o3_value}")

# Convert list to pandas DataFrame
df = pd.DataFrame(data, columns=["Day", "O3"])

# Export DataFrame to CSV file
df.to_csv("o3_data_pandas.csv", index=False)

# Release
actions.release().perform()

# Close browser
driver.quit()
```
